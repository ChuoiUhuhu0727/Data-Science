```python
# divide into categorical and numerical groups
num_cols = ['amount', 'time', 'customer_age', 'merchant_rating']
cat_cols = ['transaction_type', 'location'] # we will choose one-hot encoding
```

```python
# output of get_dummies is Dataframe
df_cat = pd.get_dummies(df[cat_cols])
```

```python
# combine categorical and numerical variables
features = pd.concat([df[num_cols], df_cat], axis=1)
```

```python
# turn boolean columns into
# int data types
for col in features.columns:
  if features[col].dtypes == 'bool':
    features[col] = features[col].astype(int)

features
```

```python
# calculate each seperate value how far from the mean
# in the units of std
scale = StandardScaler()
f_scale = scale.fit_transform(features)
f_scale
```

```python
# build train/test sets
x_train, x_test, y_train, y_test = train_test_split(features, df['is_fraud'], test_size=0.3, random_state=42)
```

```python
# calculate each seperate value how far from the mean
# in the units of std
scale = StandardScaler()
scale.fit(x_train) # learn mean/std from train only

# transform on both train and test sets
x_train_scaled = scale.transform(x_train)
x_test_scaled = scale.transform(x_test) # apply same scale
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
# vote nearest neighbors
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_scaled, y_train)
```

```python
y_pred = knn.predict(x_test_scaled)
print("Accuracy score is: ", accuracy_score(y_test, y_pred))
print("Classification Report is: \n", classification_report(y_test, y_pred))
```
