```python
# step 3: lpad and inspect data
from sklearn.datasets import load_iris
data = load_iris()
x, y = data.data, data.target
print(pd.DataFrame(x, columns=data.feature_names).head())
print(pd.Series(y).value_counts())
```

```python
# step 5: train/test split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.3, random_state=42)
```

```python
# step 7: make predictions & evaluate
y_pred = knn.predict(x_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Report: \n", classification_report(y_test, y_pred))
```

```python
# step 8: advanced tip
from sklearn.model_selection import GridSearchCV

params = {'n_neighbors': [3,5,7,9], 'weights': ['uniform', 'distance']}
grid = GridSearchCV(KNeighborsClassifier(), params, cv=5)
grid.fit(x_train, y_train)
print("Best params: ", grid.best_params_)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score, classification_report
```

```python
# load data
from sklearn.datasets import load_iris

data = load_iris()
data
```

```python
x, y = data.data, data.target
print(pd.DataFrame(x, columns=data.feature_names).head)
print(pd.Series(y).value_counts())
```

```python
# preprocess data
scale = StandardScaler()
scale.fit_transform(x)
```

```python
# build training and testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
# vote the nearest neighbors
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
# after training, we will test and compare results
y_pred = knn.predict(x_test)
```

```python
# compare before and after
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Classification report: \n", classification_report(y_test, y_pred))
```

```python
# find optimal path
from sklearn.model_selection import GridSearchCV

params = {'n_neighbors': [3,5,7], 'weights': ['distance', 'uniform']}
grid = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=params, cv=5)
grid.fit(x_train, y_train)
print("Grid best params: ", grid.best_params_)
```
